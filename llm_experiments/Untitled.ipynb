```python
import pandas as pd
import numpy as np

# Load the CSV
df = pd.read_csv('results/qwen_math/qwen_math_math_base_power_samp_batch1_mcmc10_temp0.25_seed100_20251105_204148.csv')
```

```python
# See first few rows
df.head()
```

```python
# Compare all answers
df[['naive_answer', 'std_answer', 'mcmc_answer', 'correct_answer']]
```

```python
# Check correctness (convert to string to handle formatting differences)
df['naive_correct'] = df['naive_answer'].astype(str) == df['correct_answer'].astype(str)
df['std_correct'] = df['std_answer'].astype(str) == df['correct_answer'].astype(str)
df['mcmc_correct'] = df['mcmc_answer'].astype(str) == df['correct_answer'].astype(str)
```

```python
naive_acc = df['naive_correct'].sum()
std_acc = df['std_correct'].sum()
mcmc_acc = df['mcmc_correct'].sum()
total = len(df)

print(f"Accuracy:")
print(f"  Naive (temp=0.25): {naive_acc}/{total} = {100*naive_acc/total:.1f}%")
print(f"  Standard sampling: {std_acc}/{total} = {100*std_acc/total:.1f}%")
print(f"  MCMC power samp:   {mcmc_acc}/{total} = {100*mcmc_acc/total:.1f}%")
```
